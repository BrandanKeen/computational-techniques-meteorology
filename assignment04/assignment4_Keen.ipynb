````python
# Importing Modules
import numpy as np
import matplotlib.pyplot as plt

# Definition to calculate percent change and total raw change
def percent_change_high_low(continental_population,country_names):
    # Defining empty list for percent changes and total changes
    percent_changes = []
    total_population_changes = []
   
    for country_pop in continental_population:
        # Calculating percent difference and total change
        difference = np.diff(country_pop,n=1)
        percent_diff = (difference / country_pop[:-1]) * 100
        total_change = country_pop[-1] - country_pop[0]
        
        # Appending percent changes to list
        percent_changes.append(percent_diff)

        #Appending total changes to list
        total_population_changes.append(total_change)
        
    # Appending percent changes to their respective countries
    country_percent_changes = zip(percent_changes,country_names) 
    country_percent_changes = list(country_percent_changes)
    
    # Appending percent changes to their respective countries
    total_population_changes = zip(total_population_changes,country_names)
    total_population_changes = list(total_population_changes)
    return country_percent_changes,total_population_changes

# Defining Path and File name
path = '/home/jovyan/MET3601/FALL-2024/assignments/'
file = 'world_population_data_1950_2020.csv' 

# Reading in population data
country_names = np.genfromtxt(path+file, delimiter=',', dtype=np.str_, usecols= 0, skip_header=1)
year_intervals = np.genfromtxt(path+file, delimiter=',', dtype=np.int_, usecols= range(1,16),max_rows=1)
all_populations = np.genfromtxt(path+file, delimiter=',', dtype=np.int_, usecols= range(1,16), skip_header=1)

# Grouping countries and populations based on continents
# Continent Populations
americas_pop = all_populations[12:18]
europe_pop = all_populations[9:12]
asia_pop = all_populations[7:9]
africa_pop = all_populations[1:8]
oceania_pop = all_populations[18:20]

# Country names based on continents
americas_countries = country_names[12:18]
europe_countries = country_names[9:12]
asia_countries = country_names[7:9]
africa_countries = country_names[0:7]
oceania_countries = country_names[18:20]

# Summation of continent populations for each year interval
# Source: https://numpy.org/doc/stable/reference/generated/numpy.sum.html#numpy.sum
total_americas_pop = (np.sum(americas_pop, axis=0))*1000
total_europe_pop = (np.sum(europe_pop, axis=0))*1000
total_asia_pop = (np.sum(asia_pop, axis=0))*1000
total_africa_pop = (np.sum(africa_pop, axis=0))*1000
total_oceania_pop = (np.sum(oceania_pop, axis=0))*1000


# Plotting country population data
plt.figure(1,figsize=(8,8))
plt.plot(year_intervals, total_americas_pop, label='Americas')
plt.plot(year_intervals, total_europe_pop, label='Europe')
plt.plot(year_intervals, total_asia_pop, label='Asia')
plt.plot(year_intervals, total_africa_pop, label='Africa')
plt.plot(year_intervals, total_oceania_pop, label='Oceania')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Total Population by Continent (1950-2020)',fontweight='bold')
plt.legend()

# Making y-axis logarithmic
# Source: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html
plt.yscale('log')

# Adding a figure caption
# Code adapted from sources: 
# https://stackoverflow.com/questions/34010205/adding-caption-below-x-axis-for-a-scatter-plot-using-matplotlib
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figtext.html
plt.figtext(0.1, 0.01, '\nFigure 1: ', fontstyle = 'italic', fontweight = 'bold', horizontalalignment='center',wrap=True,
            verticalalignment='top', fontsize=10)
plt.figtext(0.53, 0.01, '\nPopulation growth by continent from 1950 to 2020, depicted using a logarithmic scale.', fontstyle = 'italic',
           wrap=True, horizontalalignment='center',verticalalignment='top', fontsize=10)
# Adjusting layout
# Source: https://matplotlib.org/stable/users/explain/axes/tight_layout_guide.html
plt.tight_layout()

plt.savefig('HW4_plot1_Keen.png',bbox_inches='tight')
plt.show()


# Calling function to calculate all population percent changes and total changes
americas_percent_changes,americas_total_changes = percent_change_high_low(americas_pop,americas_countries)
europe_percent_changes,europe_total_changes = percent_change_high_low(europe_pop,europe_countries)
asia_percent_changes,asia_total_changes = percent_change_high_low(asia_pop,asia_countries)
africa_percent_changes,africa_total_changes = percent_change_high_low(africa_pop,africa_countries)
oceania_percent_changes,oceania_total_changes = percent_change_high_low(oceania_pop,oceania_countries)

# Combining all country population percent changes
all_country_percent_changes = (americas_percent_changes + europe_percent_changes + 
                               asia_percent_changes + africa_percent_changes + oceania_percent_changes)

# Combining all country total population changes
all_country_total_changes = (americas_total_changes + europe_total_changes + 
                               asia_total_changes + africa_total_changes + oceania_total_changes)

# Printing percent changes for all countries for all 5-year intervals
print('\n\n Percent Changes in Population per Country\n')
for percent_changes, country_name in all_country_percent_changes:
    print(f'{country_name}:')
    for year, change in zip(year_intervals, percent_changes):
        print(f'{year}-{year+5}: {change:.2f}%')
    print('\n') 
    
# Creating a formatted list of 5-year intervals for bar graph x-axes
year_ranges = []
for year in year_intervals:
    if (year < 2000)&(year!=1995) :
        start_year = year - 1900
        end_year = (year + 5) - 1900
    elif year == 1995:
        start_year = 95
        end_year = 00
    else:
        start_year = year - 2000
        end_year = (year + 5) - 2000
    
    year_range = f"{start_year:02d}'-{end_year:02d}'"
    year_ranges.append(year_range)
    
    if year == 2015:
        break
   
# Defining U.S.A., Germany, China, India datasets 
united_states = all_country_percent_changes[5]
germany = all_country_percent_changes[8]
china = all_country_percent_changes[9]
india = all_country_percent_changes[10]

# Plotting U.S.A., Germany, China, India percent changes
plt.figure(2,figsize = (12,12))

# United States
ax1 = plt.subplot(2, 2, 1)
ax1.bar(year_ranges,united_states[0],color='red',edgecolor='black')
ax1.set_ylim(0,10)
ax1.tick_params(axis='x',labelsize = 6)
ax1.set_xlabel('Years')
ax1.set_ylabel('Percent Change (%)')
ax1.set_title('\nUnited States')


# Germany
ax2 = plt.subplot(2, 2, 2)
ax2.bar(year_ranges,germany[0],color='green',edgecolor='black')
ax2.set_ylim(-1.2,5)
ax2.axhline(y=0, color='black', linewidth=1)
ax2.tick_params(axis='x',labelsize = 6)
ax2.set_xlabel('Years')
ax2.set_ylabel('Percent Change (%)')
ax2.set_title('\nGermany')


# China
ax3 = plt.subplot(2, 2, 3)
ax3.bar(year_ranges,china[0],color='gold',edgecolor='black')
ax3.set_ylim(0,15)
ax3.tick_params(axis='x',labelsize = 6)
ax3.set_xlabel('Years')
ax3.set_ylabel('Percent Change (%)')
ax3.set_title('\nChina')

# India
ax4 = plt.subplot(2, 2, 4)
ax4.bar(year_ranges,india[0],color='darkorange',edgecolor='black')
ax4.set_ylim(0,13)
ax4.tick_params(axis='x',labelsize = 6)
ax4.set_xlabel('Years')
ax4.set_ylabel('Percent Change (%)')
ax4.set_title('\nIndia')


# Adding title to figure
# Source:https://how2matplotlib.com/figsuptitle.html#:~:text=When%20creating%20visualizations%20using%20Matplotlib,%20it%20is%20often
plt.suptitle('Percent Change in Population by Country (1950 - 2020)\n', fontsize=16,fontweight = 'bold', horizontalalignment='center')

# Adding a figure caption
# Code adapted from sources: 
# https://stackoverflow.com/questions/34010205/adding-caption-below-x-axis-for-a-scatter-plot-using-matplotlib
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figtext.html
plt.figtext(0.06, 0.01, '\nFigure 2: ', fontstyle = 'italic',fontweight='bold', horizontalalignment='center',wrap=True,
            verticalalignment='top', fontsize=10)
plt.figtext(0.4, 0.01, '\nPercent change in population over 5 year intervals for U.S., Germany, China, and India from 1950 to 2020.',
            fontstyle = 'italic', wrap=True, horizontalalignment='center',verticalalignment='top', fontsize=10)
# Adjusting layout
# Source: https://matplotlib.org/stable/users/explain/axes/tight_layout_guide.html
plt.tight_layout()

# Saving figure
plt.savefig('HW4_plot2_Keen.png',bbox_inches='tight')
print('\n\n')
plt.show()

# Determining 5 countries with the highest and lowest population growths
all_country_total_changes = np.array(all_country_total_changes)
total_pop_changes = (all_country_total_changes[:,0]).astype(int)
sorted_country_pop_changes = np.argsort(total_pop_changes)
sorted_country_changes = all_country_total_changes[sorted_country_pop_changes]

# Setting indices and list numbers for looping to print sorted pop values
i = 19 # For printing top 5
x = 1
j = 0 # For printing bottom 5
y = 1


# Printing top 5 growth countries
print('\n\nThe top five population growth countries:\n')

for country in sorted_country_changes:
    country_name = sorted_country_changes[i, 1]
    pop_change = sorted_country_changes[i, 0]
    print(f'{x}. {country_name} - population grew by {pop_change}')
    # Updating index and list number
    i-=1
    x+=1
    # Ending the loop after the 5th country
    if x == 6:
        break

# Printing bottom 5 growth countries
print('\n\nThe bottom five population growth countries:\n')

for country in sorted_country_changes:
    country_name = sorted_country_changes[j, 1]
    pop_change = sorted_country_changes[j, 0]
    print(f'{y}. {country_name} - population grew by {pop_change}')
    # Updating index and list number
    j+=1
    y+=1
    # Ending the loop after the 5th country
    if y == 6:
        break
````

````python
# Importing Modules
import numpy as np
import datetime as dt 
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Defining Path and File name
path = '/home/jovyan/MET3601/FALL-2024/assignments/'
file1 = 'sea_ice_extent_daily.gsfs.nasateam.1978-2021.csv' 
file2 = 'avg_sea_ice_extent_modified.csv'

# Reading in sea ice data
daily_sea_ice_extent_doy = np.genfromtxt(path+file1, delimiter=',', dtype=np.int_, usecols= 1, skip_header=1)
daily_sea_ice_extent = np.genfromtxt(path+file1, delimiter=',', dtype=np.int_, usecols= range(3,13), skip_header=1)
daily_sea_ice_extent_dates  = np.genfromtxt(path+file1, delimiter=',', dtype=np.datetime64, usecols=0, skip_header=1)
average_sea_ice_extent = np.genfromtxt(path+file2, delimiter=',', dtype=np.float_, skip_header=2)
average_sea_ice_extent_dates = np.genfromtxt(path+file2, delimiter=',', dtype=np.datetime64, usecols=0, skip_header=2)


# Converting to Datetime Objects
# Code adapted from Dr. L's lecture notes 12
daily_dates = daily_sea_ice_extent_dates.astype('datetime64[D]').astype('O')

# Defining start and end dates to subset
start_date = np.datetime64('2017-01-01')
end_date = np.datetime64('2021-12-31')

subset = (daily_dates >= start_date)&(daily_dates <= end_date)

# Defining subset sea ice extent and dates for 2017-2021
daily_sea_ice_extent_2017_2021 = daily_sea_ice_extent[subset]
dates_2017_2021 = daily_dates[subset]
doy = daily_sea_ice_extent_doy[subset]

# Stacking subset arrays horizontally
daily_sea_ice_extent_2017_2021 = np.column_stack((dates_2017_2021,doy,daily_sea_ice_extent_2017_2021))


# Defining year,doy portions of datetime object in subset array
years = np.array([date.year for date in daily_sea_ice_extent_2017_2021[:, 0]])
doy = np.array(daily_sea_ice_extent_2017_2021[:, 1])

# Zipping arrays to iterate through together
daily_sea_ice_extent_2017_2021 = zip(years,daily_sea_ice_extent_2017_2021)

# Empty Lists to append values to by year
year_2017 = []
year_2018 = []
year_2019 = []
year_2020 = []
year_2021 = []
summer_solstice_dates = []
winter_solstice_dates = []

 # Iterating through array, appending values to lists by year and pulling solstice days
for years,row in daily_sea_ice_extent_2017_2021:
    doy = row[1]
    if years == 2017:
        year_2017.append(row)
        if doy == 172:    # doy for Summer Solstice          
            summer_solstice_dates.append(row)
        elif doy == 355:  # doy for Winter Solstice
            winter_solstice_dates.append(row)
    elif years == 2018:
        year_2018.append(row)
        if doy == 172:
            summer_solstice_dates.append(row)
        elif doy == 355:
            winter_solstice_dates.append(row)
    elif years == 2019:
        year_2019.append(row)
        if doy == 172:
            summer_solstice_dates.append(row)
        elif doy == 355:
            winter_solstice_dates.append(row)
    elif years == 2020:
        if doy != 60:                # Eliminating leap day February 29th from appeneded rows for 2020
            year_2020.append(row)
        if doy == 173:
            summer_solstice_dates.append(row)
        elif doy == 356:
            winter_solstice_dates.append(row)
    elif years == 2021:
        year_2021.append(row)
        if doy == 172:
            summer_solstice_dates.append(row)
        elif doy == 355:
            winter_solstice_dates.append(row)


# Combining Summer and Winter Solstice dates
combined_solstice_dates = summer_solstice_dates + winter_solstice_dates   

# Converting lists to arrays
year_2017 = np.array(year_2017)
year_2018 = np.array(year_2018)
year_2019 = np.array(year_2019)
year_2020 = np.array(year_2020)
year_2021 = np.array(year_2021)
combined_solstice_dates = np.array(combined_solstice_dates)

# Defining date column in arrays
dates_year_2017 = year_2017[:, 0].astype('datetime64[D]')
dates_year_2018= year_2018[:, 0].astype('datetime64[D]')
dates_year_2019= year_2019[:, 0].astype('datetime64[D]')
dates_year_2020= year_2020[:, 0].astype('datetime64[D]')
dates_year_2021= year_2021[:, 0].astype('datetime64[D]')
combined_solstice_dates = combined_solstice_dates[:,0].astype('datetime64[D]')

# Defining Total Arctic sea ice extent and baseline statistics
totalarc_2017 = year_2017[:,2]/1e6       # Scaling values to match baseline data 
totalarc_2018= year_2018[:,2]/1e6
totalarc_2019= year_2019[:,2]/1e6
totalarc_2020= year_2020[:,2]/1e6
totalarc_2021= year_2021[:,2]/1e6
baseline_1981_2010= average_sea_ice_extent[:,1]
percentile_10th = average_sea_ice_extent[:,3]
percentile_90th = average_sea_ice_extent[:,7]

# Plotting daily sea ice extent for years 2017-2021
plt.figure(figsize=(20,12))

# Formatting x-axis to display month-day-year
# Code adapted from Source: https://stackoverflow.com/questions/14946371/editing-the-date-formatting-of-x-axis-tick-labels
format = mdates.DateFormatter('%m-%d-%Y')
plt.gca().xaxis.set_major_formatter(format)

# Setting the x-axis to display ticks for each solstice
plt.gca().set_xticks(combined_solstice_dates)
plt.xticks(rotation=45)

# Plotting daily sea ice extent data
plt.plot(dates_year_2017, totalarc_2017, label='2017',linewidth = 2)  
plt.plot(dates_year_2018, totalarc_2018, label= '2018',linewidth = 2)
plt.plot(dates_year_2019, totalarc_2019, label='2019',linewidth = 2)  
plt.plot(dates_year_2020, totalarc_2020, label= '2020',linewidth = 2)
plt.plot(dates_year_2021, totalarc_2021, label='2021',linewidth = 2)  

# Plotting 1981-2010 baseline
plt.plot(dates_year_2017,baseline_1981_2010, color='black', linestyle= 'dashed',label='1981-2010 Mean ')
plt.plot(dates_year_2018,baseline_1981_2010, color='black', linestyle= 'dashed')
plt.plot(dates_year_2019,baseline_1981_2010, color='black', linestyle= 'dashed')
plt.plot(dates_year_2020,baseline_1981_2010, color='black', linestyle= 'dashed')
plt.plot(dates_year_2021,baseline_1981_2010, color='black', linestyle= 'dashed')

# Shading region between 10th and 90th percentiles
plt.fill_between(dates_year_2017, percentile_10th, percentile_90th, color='grey', alpha=0.3, label= '10%-90% Range')
plt.fill_between(dates_year_2018, percentile_10th, percentile_90th, color='grey', alpha=0.3)
plt.fill_between(dates_year_2019, percentile_10th, percentile_90th, color='grey', alpha=0.3)
plt.fill_between(dates_year_2020, percentile_10th, percentile_90th, color='grey', alpha=0.3)
plt.fill_between(dates_year_2021, percentile_10th, percentile_90th, color='grey', alpha=0.3)

# Detailing graph
plt.xlabel('Date')
plt.ylabel('Sea Ice Extent ($10^6$ km)')
plt.title('Total Arctic Sea Ice Extent (2017-2021)')
plt.legend()
plt.grid(True)
plt.figtext(0.05, 0.01, '\n\nFigure 3:',fontstyle = 'italic',fontweight='bold', horizontalalignment='center',wrap=True, verticalalignment='top', fontsize=10)
plt.figtext(0.207, 0.01, '\n\nTimeseries of total Arctic sea ice extent from 2017 to 2021 (multicolor curve).',
            fontstyle = 'italic', wrap = True, horizontalalignment='center',verticalalignment='top', fontsize=10)
plt.figtext(0.533, 0.01, '\n\n 1981-2010 mean depicted by black dashed line, 10th to 90th percentile range shown by the shaded region.',
            fontstyle = 'italic', wrap = True, horizontalalignment='center',verticalalignment='top', fontsize=10)
plt.figtext(0.859, 0.01, '\n\n Summer and Winter Solstice dates labeled to highlight seasonal fluctuations. ',
            fontstyle = 'italic', wrap = True, horizontalalignment='center',verticalalignment='top', fontsize=10)
plt.figtext(0.201, 0.01, '\n\n\n The data was sourced from the National Snow and Ice Data Center (NSIDC).',
            fontstyle = 'italic', wrap = True, horizontalalignment='center',verticalalignment='top', fontsize=10)

plt.tight_layout()
plt.savefig('HW4_plot3_Keen.png',bbox_inches='tight')
plt.show()

````
